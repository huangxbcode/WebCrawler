当你跟踪的微博大v发新微博时，自动点赞，自动回复。

# 查看提示

•	问题拆解提示

•	该问题可以拆解为以下若干子问题：

•	1.如何访问该用户的页面？

•	2.如何发现未点赞的微博？

•	3.如何自动点击按钮？如何自动点击评论、编辑、发布评论

•	4.如何持续监测？

•	问题解决提示

•	1.利用 selenium 库中的 webdriver，先访问微博的登录页面，手动登录，然后再访问某特定用户的微博主页。

•	2.将微博的时间标签当作一条微博的唯一标识，将点赞过的微博存入 list 中，如果一条新的时间标签不在 list 中，则说明需要点赞。

•	3.利用 click 函数，实现模拟点击。

•	4.利用 while True 实现持续监测。

•	注：由于微博的限制，点赞过快可能会导致页面变化，引起程序崩溃。建议点赞间隔设置的长一些。


```python
# coding:utf-8
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
# 打开Chrome浏览器
def start_chrome():
    # windows系统下，executable_path='./chromedriver.exe'
#     driver = webdriver.Chrome(executable_path='./chromedriver')
    print("start_chrome...")
    driver = webdriver.Chrome()
    driver.start_client()
    driver.set_window_size(1500, 1500)
    return driver
# 打开登录界面
def login(driver, url):
    print("login in 60s...")
    driver.get(url)
    time.sleep(60)
# 点赞函数
def click_like(driver, url, have_clicked):
    print("click_like...start")

    # 获取 https://weibo.com/bgsxy 每条微博
#     weibo_sel = '#Pl_Official_MyProfileFeed__21 > div > div' # 每条微博的div标签
#     time_sel = 'div.WB_feed_detail > div.WB_detail > div.WB_from.S_txt2 > a:nth-child(1)' # 一条微博的时间
#     button_sel = 'div.WB_feed_handle > div > ul > li:nth-child(4) > a > span > span > span' # 按钮位置
    
    # 获取 https://weibo.com/ideamagic 每条微博
    weibo_sel = '#Pl_Official_MyProfileFeed__21 > div > div' # 每条微博的div标签
    time_sel = 'div.WB_feed_detail > div.WB_detail > div.WB_from.S_txt2 > a:nth-child(1)' # 一条微博的时间
    like_sel = 'div.WB_feed_handle > div > ul > li:nth-child(4) > a > span > span > span' # 点赞 按钮位置
    
    comment_sel = 'div.WB_feed_handle > div > ul > li:nth-child(3) > a > span > span > span' # comment按钮位置
    
    edit_input_sel = 'div.WB_feed_repeat > div > div > div.WB_feed_publish > div.WB_publish > div.p_input' # comment input
    edit_input_id_sel = '' # comment input id
    button_input_sel   = 'div.WB_feed_repeat > div > div > div.WB_feed_publish > div.WB_publish > div.p_opt > div.btn.W_fr > a' # 评论
    
    driver.get(url)
    # 等待页面加载
    time.sleep(6)
    # 获取 https://weibo.com/bgsxy 每条微博的div标签
#     weibo_sel = '#Pl_Official_MyProfileFeed__21 > div > div'
    weibo_elems = driver.find_elements_by_css_selector(weibo_sel)
    print("click_like...find_elements_by_css_selector")
    for w in weibo_elems:
        # 获取 https://weibo.com/bgsxy  一条微博的时间
#         time_sel = 'div.WB_feed_detail > div.WB_detail > div.WB_from.S_txt2 > a:nth-child(1)'
        weibo_time = w.find_elements_by_css_selector(time_sel)
        # 由于页面结构的原因，有的标签中没有时间信息，所以此处加了一个条件判断
        if len(weibo_time) == 1:
            weibo_time = weibo_time[0].text
            print("weibo_time:", weibo_time)
        if weibo_time not in have_clicked:
#             button_sel = 'div.WB_feed_handle > div > ul > li:nth-child(4) > a > span > span > span'
            button_like = w.find_elements_by_css_selector(like_sel)
            click_result = False
    
            # 同样由于页面结构的原因，此处加了一个条件判断
            if len(button_like) == 1:
                button_like = button_like[0]
                # 微博页面上的“私信聊天”按钮，有时候会挡住点赞按钮，遇到这个问题的时候，
                # 我们先捕获异常，然后利用“下方向键”让页面自动滚动一点点，露出点赞按钮
                try:
                    print("now is ", time.asctime( time.localtime(time.time()) )," 点赞...")
                    button_like.click()
                    click_result = True
                except Exception:
#                     html_page = driver.find_element_by_tag_name('html')
#                     html_page.send_keys(Keys.DOWN)
                    print("点赞失败")
#                     print(html_page)
#                     button.click()
                    click_result = False
                have_clicked.append(weibo_time)
                # 微博的点赞速度过快会弹出提醒，所以时间间隔设置的长一点
                time.sleep(5)
                
            if click_result :
                # get comment location and click
                button_comment = w.find_elements_by_css_selector(comment_sel)
                                
                if len(button_comment) == 1:
                    button_comment = button_comment[0]
                    try:
                        print("now is ", time.asctime( time.localtime(time.time()) )," 进入评论...")
                        button_comment.click()
                        # wait 10s to load comment page
                        time.sleep(6)
                    except Exception:
    #                     html_page = driver.find_element_by_tag_name('html')
    #                     html_page.send_keys(Keys.DOWN)
                        print("进入评论失败")

                    # input "给掌柜点赞" and click “评论”
                    print("发送评论--输入点赞评论内容")
                    w.find_element_by_class_name('W_input').send_keys("给掌柜点赞！")  
                    time.sleep(1)
                    button_input = w.find_elements_by_css_selector(button_input_sel)
                    button_input = button_input[0]
                    try:
                        print("发送评论--给掌柜点赞")
                        button_input.click()
                    except Exception:
    #                     html_page = driver.find_element_by_tag_name('html')
    #                     html_page.send_keys(Keys.DOWN)
                        print("发送评论失败")

                    # time.sleep(10)



login_url = 'https://weibo.com/1405717072'
# weibo_url = 'https://weibo.com/bgsxy' # 办公室小野
weibo_url = 'https://weibo.com/ideamagic' # 常程
# 用时间做标识，来判断一条微博是否点赞过
have_clicked = []
driver = start_chrome()
login(driver, login_url)

cont = "c"
while cont == "c":
    have_clicked = click_like(driver, weibo_url, have_clicked)
    print("now is ", time.asctime( time.localtime(time.time()) ))
#     print("sleep 2 minutes...")
#     time.sleep(1200)
    print("type c if you wang to continue...")
    cont = input()
```
